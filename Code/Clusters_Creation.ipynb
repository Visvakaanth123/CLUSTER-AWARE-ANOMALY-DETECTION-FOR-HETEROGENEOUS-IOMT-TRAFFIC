```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

pd.set_option('display.float_format', '{:.6f}'.format)
dataset = pd.read_csv("Benign_train.pcap.csv")
scalar = StandardScaler()
scaled_dataset = scalar.fit_transform(dataset)
model = KMeans(n_clusters=5)
labels = model.fit_predict(scaled_dataset)

dataset['Cluster'] = labels
centroids = pd.DataFrame(scalar.inverse_transform(model.cluster_centers_), columns=dataset.columns[:-1])
print(centroids.T)
k = dataset['Cluster'].nunique()  # number of clusters

for i in range(k):
    cluster_df = dataset[dataset['Cluster'] == i]
    cluster_df.to_csv(f"cluster_{i}.csv", index=False)
    print(f"Saved cluster {i} with {len(cluster_df)} rows")
```
